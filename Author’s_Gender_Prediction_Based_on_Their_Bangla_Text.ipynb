```python
import pandas as pd
import re

# Load the CSV data into a DataFrame
df = pd.read_csv('/content/drive/MyDrive/Author Gender Prediction/dataset.csv')

# Preprocess the data
df['pre_processed_data_1'] = df['pre_processed_data_1'].str.lower()
df['pre_processed_data_1'] = df['pre_processed_data_1'].str.replace(r'[^\w\s]', '', regex=True)

# Extract the features
def extract_features(text):
    """Extracts features from Bangla text without using spaCy."""

    # Stylistic features
    avg_sentence_length = len(text.split()) / len(text.splitlines())
    num_first_person_pronouns = len([token for token in text if token in ['আমি', 'আমার', 'আমাকে', 'আমাদের']])
    num_adjectives = len([token for token in text if re.search('(আলা|তা|ময়|মত)$', token)])
    num_adverbs = len([token for token in text if re.search('(ভাবে|তো|কি|গেলে)$', token)])

    # Linguistic features
    avg_token_length = sum([len(token) for token in text]) / len(text.split())
    num_unique_words = len(set(text.split()))
    num_stopwords = len([token for token in text if token in ['ও', 'এ', 'কে', 'যে', 'সে', 'তার', 'হয়', 'না', 'থাকে', 'যাবে']])

    return avg_sentence_length, num_first_person_pronouns, num_adjectives, num_adverbs, avg_token_length, num_unique_words, num_stopwords

# Extract the features for each book
features = []
for i in range(len(df)):
    features.append(extract_features(df['pre_processed_data_1'].iloc[i]))

# Save the extracted features to a CSV file
df_features = pd.DataFrame(features, columns=['avg_sentence_length', 'num_first_person_pronouns', 'num_adjectives', 'num_adverbs', 'avg_token_length', 'num_unique_words', 'num_stopwords'])
df_features.to_csv('bangla_book_features.csv', index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, matthews_corrcoef

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/Author Gender Prediction/dataset.csv')

# Define the pronouns
pronouns = ['আমি', 'তুমি', 'আপনি', 'সে', 'এ', 'ও','আমার','আপনার','তার','সেই','কার','কেউ','কোন','কত',
            'কী','সব','অনেকে','একজন', 'যে','যা']

# Calculate relative pronoun frequencies
for pronoun in pronouns:
    data[f'rel_{pronoun}'] = data[pronoun] / 4000  # 'total_words' is the total word count in the segment

# Define features and target
X = data[[f'rel_{pronoun}' for pronoun in pronouns]]
y = data['gender']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Define hyperparameter grid
param_grid = {
    'C': [0.1, 1.0, 10.0],
    'gamma': ['scale', 'auto']
}

# Create the SVM classifier
svm = SVC(random_state=42)

# Perform grid search to find the best hyperparameters
grid_search = GridSearchCV(svm, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Train the model with the best hyperparameters
best_svm = SVC(random_state=42, **best_params)
best_svm.fit(X_train, y_train)

# Make predictions and evaluate the model
y_pred = best_svm.predict(X_test)

# Perform 5-fold cross-validation
cross_val_scores = cross_val_score(best_svm, X, y, cv=5)

# Print the performance metrics for each fold
print("Cross-validation scores:", cross_val_scores)

# Calculate the average performance across all folds
avg_score = cross_val_scores.mean()
print("Average score:", avg_score)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, pos_label='m')
recall = recall_score(y_test, y_pred, pos_label='m')
f1 = f1_score(y_test, y_pred, pos_label='m')
mcc = matthews_corrcoef(y_test, y_pred)
confusion_svm = confusion_matrix(y_test, y_pred)

# Print the results
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("MCC:", mcc)
print(classification_report(y_test, y_pred))
print("Confusion Matrix for SVM:")
print(confusion_svm)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, matthews_corrcoef
from sklearn.feature_selection import SelectKBest, f_classif

# Define the pronouns
pronouns = ['আমি', 'তুমি', 'আপনি', 'সে', 'এ', 'ও', 'আমরা', 'তোমরা', 'আপনারা', 'তারা',
            'আমার', 'তোমার', 'আপনার', 'তার', 'আমাদের', 'তোমাদের', 'আপনাদের', 'তাদের',
            'এই', 'ওই', 'সেই', 'এগুলো', 'ওগুলো', 'সেগুলো', 'কে', 'কার', 'কোন', 'কত',
            'কী', 'কেউ', 'কিছু', 'যে কেউ', 'যা কিছু', 'সব', 'অনেকে', 'কয়েকজন', 'কিছু কিছু',
            'একজন', 'যে', 'যেখানে', 'যখন', 'যা', 'একে অপর', 'পরস্পর', 'আমি নিজে', 'তুমি নিজে',
            'সে নিজে', 'আমরা নিজেরা', 'তোমরা নিজেরা', 'তারা নিজেরা', 'আমিই', 'তুমিই', 'সেইই', 'আমরাই',
            'তোমরাই', 'তারাই', 'আমাকেই', 'তোমাকেই', 'তাকেই', 'আমাদেরকেই', 'তোমাদেরকেই', 'তাদেরকেই']

# Calculate relative pronoun frequencies
def calculate_relative_pronoun_frequencies(pronouns, data):
    """Calculates the relative pronoun frequencies for a Pandas DataFrame.

    Args:
        pronouns: A list of pronouns.
        data: A Pandas DataFrame containing the text and gender labels.

    Returns:
        A Pandas DataFrame with the relative pronoun frequencies.
    """

    X = pd.DataFrame()
    for pronoun in pronouns:
        X[f'rel_{pronoun}'] = data[pronoun] / 2000

    return X

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/Author Gender Prediction/dataset.csv')

# Calculate relative pronoun frequencies
X = calculate_relative_pronoun_frequencies(pronouns, data)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, data['gender'], test_size=0.2, random_state=42)

# Define a range of values for k
k_values = list(range(1, 63))  # You can adjust this range as needed

best_k = None
best_aggregate_score = -1  # Initialize with a lower value, we want to maximize this score

for k in k_values:
    # Select the top k most predictive pronouns
    selector = SelectKBest(f_classif, k=k)
    selector.fit(X_train, y_train)
    X_train_reduced = selector.transform(X_train)
    X_test_reduced = selector.transform(X_test)

    # Create the Decision Tree model
    decision_tree = DecisionTreeClassifier(random_state=42)

    # Fit the model to the reduced training data
    decision_tree.fit(X_train_reduced, y_train)

    # Make predictions on the test set
    y_pred = decision_tree.predict(X_test_reduced)

    # Calculate the metrics of interest
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, pos_label='m')
    precision = precision_score(y_test, y_pred, pos_label='m')
    recall = recall_score(y_test, y_pred, pos_label='m')
    mcc = matthews_corrcoef(y_test, y_pred)

    # Compute an aggregate score (e.g., weighted average of metrics)
    # You can adjust the weights based on your preferences
    aggregate_score = (
        0.2 * accuracy +
        0.2 * f1 +
        0.1 * precision +
        0.1 * recall +
        0.2 * mcc
    )

    # Check if the current k value gives a better aggregate score
    if aggregate_score > best_aggregate_score:
        best_aggregate_score = aggregate_score
        best_k = k

# Print the best k value
print("Best k:", best_k)
```
